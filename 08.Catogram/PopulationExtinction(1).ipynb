```python
df = pd.read_csv(f'{data_folder}/population_2020.csv', encoding = 'EUC-KR')
print(df.shape)
df.head()
```

```python
df['시군구'][df.행정구.notnull()] = df.행정구[df.행정구.notnull()]
df.head(266).tail()
```

```python
df.head()
```

```python
df = df[df.시군구 != '소계']
print(df.shape)
df.head()
```

```python
df['20 ~ 39세'] = df['20 - 24세'] + df['25 - 29세'] + df['30 - 34세'] + df['35 - 39세']
df['65세 이상'] = df['65 - 69세'] + df['70 - 74세'] + df['75 - 79세'] + df['80세 이상']
df = df[['광역시도', '시군구', '구분', '인구수', '20 ~ 39세', '65세 이상']]
df.head()
```

```python
pop = pd.pivot_table(
    df,
    index = ['광역시도', '시군구'],
    columns = ['구분'],
    values = ['인구수', '20 ~ 39세', '65세 이상']                     
)
print(pop.shape)
pop.head()
```

```python
pop['소멸 비율'] = pop['20 ~ 39세', '여자'] / pop['65세 이상', '계']
pop.head()
```

```python
pop['소멸 위기 지역'] = pop['소멸 비율'] < 0.5
pop['소멸 위기 고위험 지역'] = pop['소멸 비율'] < 0.2
pop.head()
```

```python
crisis_region = pop[pop['소멸 위기 지역']].index.get_level_values(1)
print(len(crisis_region))
print(crisis_region)
```

```python
high_crisis_region = pop[pop['소멸 위기 고위험 지역']].index.get_level_values(1)
print(len(high_crisis_region))
print(high_crisis_region)
```

```python
pop.columns.get_level_values(0)
```

```python
for i in range(len(pop.columns.get_level_values(0))):
    print(pop.columns.get_level_values(0)[i] + pop.columns.get_level_values(1)[i])
```

```python
tmp_col = [
           pop.columns.get_level_values(0)[i] + \
           pop.columns.get_level_values(1)[i]
           for i in range(len(pop.columns.get_level_values(0)))
]
pop.columns = tmp_col
pop.reset_index(inplace = True)
pop['시군구'] = pop.시군구.apply(lambda x: x.strip())
pop.head()
```

```python
pop.to_csv(f'{data_folder}/PreprocessingComplete.csv', encoding = 'EUC-KR', index = False)
```
