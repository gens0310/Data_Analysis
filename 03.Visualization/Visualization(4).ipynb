## 04. Matplotlib Various Plots

### 1) Bar Chart
- x 데이터가 카테고리 값인 경우

```python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.grid()
plt.show()
```

```python
np.random.seed(2021)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.grid()
plt.show()
```

#### 타이타닉 선실별 승객 수

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
s = titanic['class'].value_counts()
s
```

```python
df = titanic[['class', 'pclass']].groupby('class').count()
df
```

```python
y = s.values
x = np.arange(len(y))
xlabel = s.index.to_list()
plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel("선실")
plt.ylabel("승객 수")
plt.grid()
plt.show()
```

```python
y = df.pclass.values
x = np.arange(len(y))
xlabel = df.index.to_list()
plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel("선실")
plt.ylabel("승객 수")
plt.grid()
plt.show()
```

```python
y = df.pclass.values
x = np.arange(len(y))
xlabel = df.index.to_list()
plt.title('타이타닉 선실별 승객 수')
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.xlabel("선실")
plt.ylabel("승객 수", rotation=0, labelpad=15)
plt.grid()
plt.show()
```

### 2) Pie Chart
- 카테고리 별 값의 상대적인 비교

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10] # 합이 100
# colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, # colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10] # 합이 100
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart", fontsize=15)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

#### Anscombe Quartet
- 선형회귀선 : y = 3.0 + 0.5 * x

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

##### Dataset I 산점도와 회귀선 그리기

```python
ds1 = ans[ans.dataset == 'I']
ds1
```

```python
plt.scatter(ds1.x, ds1.y)
plt.plot([2, 16], [4, 11], c='r', lw=2)
plt.title('Dataset I')
plt.grid()
plt.show()
```

##### 4개의 데이터셋

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i // 2, i % 2] # [row, col]
    ax.scatter(ds.x, ds.y)
    ax.grid()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6), sharex=True, sharey=True)
for i, roman in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[i // 2, i % 2] # [row, col]
    ax.scatter(ds.x, ds.y)
    ax.plot([3, 20], [4.5, 13], 'r-', lw=2) # y = 0.5 * x + 3
    ax.set_title(f'Dataset {roman}')
    ax.grid()
fig.suptitle('Anscombe Quartet', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
X = digits.images[0]
X
```

```python
digits.images.shape
```

```python
plt.title("mnist digits; 0")
plt.imshow(X, interpolation='nearest', cmap=plt.cm.bone_r)
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.show()
```

```python
methods = [
    None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
    'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
    'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos'
]
fig, axes = plt.subplots(3, 6, figsize=(12, 6),
                         subplot_kw={'xticks': [], 'yticks': []})
for ax, interp_method in zip(axes.flat, methods):
    ax.imshow(X, cmap=plt.cm.bone_r, interpolation=interp_method)
    ax.set_title(interp_method)
plt.show()
```

### 6) Box Plot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
setosa = iris[iris.species == 'setosa']
plt.figure(figsize=(8, 10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
             setosa.petal_length, setosa.petal_width],
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.title('Setosa 품종의 각 Feature의 Box Plot')
plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 8), sharey=True)
for i, species in enumerate(iris.species.unique()):
    sub = iris[iris.species == species]
    ax = axes[i] # [row, col]
    ax.boxplot([sub.sepal_length, sub.sepal_width,
             sub.petal_length, sub.petal_width],
            labels=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
    ax.set_title(species)
    ax.grid()
fig.suptitle('iris 모든 품종의 Feature별 Box Plot', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
