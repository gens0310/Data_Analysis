### Alice in Wonderland

```python
from google.colab import files
uploaded = files.upload()
textfile = list(uploaded.keys())[0]
```

```python
uploaded = files.upload()
maskfile = list(uploaded.keys())[0]
```

```python
with open(textfile) as fp:
    text = fp.read()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.add('said')
```

```python
wc = WordCloud(background_color = 'white',
               max_words = 1000,
               stopwords = stop_words)
wc.generate(text)
```

```python
type(wc.words_)
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())
for i in range(10):
    print(keys[i], ':', values[i])
```

```python
plt.figure(figsize = (10, 6))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open(maskfile))
plt.imshow(mask,
           cmap = plt.cm.gray,
           interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color = 'white',
               mask = mask,
               max_words = 1000,
               stopwords = stop_words)
wc.generate(text)
```

```python
plt.figure(figsize = (10, 12))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

### Star Wars

```python
from google.colab import files
uploaded = files.upload()
textfile = list(uploaded.keys())[0]
```

```python
uploaded = files.upload()
maskfile = list(uploaded.keys())[0]
```

```python
with open(textfile) as fp:
    text = fp.read()
text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')
```

```python
from wordcloud import STOPWORDS
stop_words = set(STOPWORDS)
stop_words.add('int')
stop_words.add('ext')
len(stop_words)
```

```python
mask = np.array(Image.open(maskfile))
wc = WordCloud(mask = mask,
               max_words = 1000,
               stopwords = stop_words,
               margin = 10,
               random_state = 2021)
wc.generate(text)
default_color = wc.to_array
```

```python
import random
def grey_color(word, font_size, position,
               orientation, random_state = None,
               **kwargs):
    return f'hsl(0, 0%, {random.randint(60, 100)}%)'
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wc.recolor(color_func = grey_color, random_state = 2021), 
           interpolation='bilinear')
plt.axis('off')
plt.show()
```
