### 한글 텍스트 전처리

```python
with open(textfile) as fp:
    text = fp.read()
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
tokens = okt.nouns(text)
tokens[:10]
```

#### 영문, 숫자 제거

```python
import re
new_tokens = []
for token in tokens:
    gift = re.sub('[A-Za-z0-9]', '', token)
    if gift:
        new_tokens.append(gift)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family = 'NanumBarunGothic')
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

#### 반복하여 불용어 정리

```python
gift = nltk.Text(new_tokens, name = '여친선물')
plt.figure(figsize = (15, 6))
gift.plot(50)
plt.show()
```

```python
stoptext = '''
선물 추천 것 가격 수 기능 제품 저 제 생각 여자
여친 여자친구 사용 요 더 사진 구매 고급 판매 때 참고
하나 한번 제작 해 용 준 디자인 거 네이버 사람 배송
중 정발 후기 개 감동 하트 커플 가능 사랑 분 머리 세대
인기 상품 직접 브랜드 선택 모양 마감 곳 등 요즘 오늘
가지 남녀 위 페이 마음 부담 향 끈 곱창 그냥 남자 협찬
전 핸드 의미 변경 도움 색상 아래 포장 처 조금 하루 정도
처리 보시 독서 꽃말 심플 인테리어 이벤트 여성 가장 때문
정보 스타일 소품 밍티풀 주소 쇼핑 링크 이니셜 기념일 질문
친구 스토어아주 할인 미쁜 스
'''
stop_words = stoptext.split()
new_tokens = [word for word in new_tokens if word not in stop_words]
new_tokens[:10]
```

```python
from wordcloud import WordCloud
from PIL import Image
```

```python
data = gift.vocab().most_common(300)
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
wc = WordCloud(
    font_path = path,
    relative_scaling = 0.2,
    background_color = 'white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize = (12, 6))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

#### 그림 위에 표시

```python
mask = np.array(Image.open(maskfile))
from wordcloud import ImageColorGenerator
image_colors = ImageColorGenerator(mask)
```

```python
wc = WordCloud(
    font_path = path,
    relative_scaling = 0.2,
    background_color = 'white',
    mask = mask,
    min_font_size = 1,
    max_font_size = 120
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(wc.recolor(color_func = image_colors),
           interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
